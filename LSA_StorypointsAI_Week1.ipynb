### Task 1: Explore schemas, nulls, duplicates

```python
def explore_dataset(file_path, name):
    print(f"\n[INFO] Exploring dataset: {name}")
    df = pd.read_excel(file_path, engine="pyxlsb")

    schema = df.dtypes
    nulls = df.isnull().sum()
    dups = df.duplicated().sum()

    print(f"\nSchema for {name}:\n{schema}")
    print(f"\nNull counts for {name}:\n{nulls}")
    print(f"\nDuplicates in {name}: {dups}")

    return df.head(5)
```

### Task 3: Transform

```python
def process_chunk(chunk, rates):
    # Standardize column names
    chunk.columns = [col.strip().lower().replace(" ", "_") for col in chunk.columns]
 # Convert timestamps
    if "txn_time" in chunk.columns:
        chunk["txn_time"] = pd.to_datetime(chunk["txn_time"], utc=True)
    else: 
        chunk["click_time"] = pd.to_datetime(chunk["click_time"], utc = True)

    # Deduplicate
    chunk = chunk.drop_duplicates()

    # Enrich transactions
    if "currency" in chunk.columns and "amount" in chunk.columns:
        chunk["amount_in_usd"] = chunk.apply(
            lambda row: row["amount"] / rates.get(row["currency"], 1)
            if pd.notnull(row["amount"]) else None,
            axis=1
        )

    return chunk
```

#### Task 4: Load — write cleaned outputs to GCS paths partitioned by ingest_date=YYYY-MM-DD/ for each dataset.

```python
def upload_to_gcs(df, bucket_name, prefix, dataset_name):
    client = storage.Client(project="winter-agility-471714-p2")
    bucket = client.bucket(bucket_name)

    ingest_date = datetime.utcnow().strftime("%Y-%m-%d")
    dest_path = f"{BPATH}/{prefix}/{dataset_name}/ingest_date={ingest_date}/{dataset_name}.parquet"

    tmp_file = f"/tmp1/{dataset_name}.parquet"
    df.to_parquet(tmp_file, index=False)
    blob = bucket.blob(dest_path)
    blob.upload_from_filename(tmp_file)
    print(f"[SUCCESS] Uploaded {dataset_name} → gs://{bucket_name}/{dest_path}")
```

```python
all_chunks = []
for chunk in pd.read_csv("transactions.csv", chunksize=50000):
    all_chunks.append(process_chunk(chunk, rates))
    upload_to_gcs(df_transactions, BUCKET_NAME, DEST_PREFIX, "transactions")

df_transactions = pd.concat(all_chunks, ignore_index=True)
```

```python
all_chunks = []
for chunk in pd.read_csv("clickstream.csv", chunksize=50000):
    all_chunks.append(process_chunk(chunk, rates))
    upload_to_gcs(df_clickstream, BUCKET_NAME, DEST_PREFIX, "clickstream")

df_transactions = pd.concat(all_chunks, ignore_index=True)
```
